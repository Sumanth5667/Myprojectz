### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs'].value_counts()
```

```python
# Create a frequency table
frequency_table = df.groupby('App')['Installs'].sum().reset_index()

# Sort the frequency table by the number of installs in descending order
frequency_table = frequency_table.sort_values(by='Installs', ascending=False)

# Get the top 10 most popular apps
top_10_apps = frequency_table.head(10)

# Print the top 10 most popular apps
print(top_10_apps)






```
