Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of 
Matplotlib

ANS:
    Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in 2D and 3D. It provides a wide range of functions and tools for visualizing data in various formats and styles. Matplotlib is widely used in the scientific and data analysis communities for creating publication-quality plots, charts, graphs, and figures.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line Plot: A basic plot where data points are connected by straight lines. It is useful for showing the trend or relationship between continuous data points over time or other continuous variables.

2. Scatter Plot: A plot that represents individual data points as dots, with each dot's position determined by the values of two variables. Scatter plots are useful for visualizing the distribution and relationship between two numerical variables.

3. Bar Plot: A plot that uses rectangular bars to represent data, with the length or height of each bar proportional to the data values. Bar plots are commonly used to compare and display categorical data.

4. Histogram: A plot that shows the distribution of a continuous variable by dividing it into bins and counting the number of occurrences in each bin. Histograms help visualize the data's underlying frequency distribution.

5. Pie Chart: A circular chart divided into slices, where each slice represents a portion of the whole data or a percentage of a category. Pie charts are suitable for displaying the relative proportions of different categories in a dataset.

These are just a few examples of the plots that can be created using Matplotlib's Pyplot module. There are many other types of plots and customization options available to create diverse visualizations.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization used to display the relationship between two numerical variables. It represents individual data points as dots on a 2D plane, with each dot's position determined by the values of the two variables being compared. One variable is plotted on the x-axis, and the other variable is plotted on the y-axis.

The main purpose of a scatter plot is to show how two variables are related to each other, whether they exhibit a pattern or trend, and to identify any potential correlations or outliers in the data. Scatter plots are particularly useful for visualizing the distribution and relationship between continuous data points.

In a scatter plot, each dot represents a single data point, and its position on the graph corresponds to the values of the two variables it represents. If the dots cluster around a straight line going from the bottom left to the top right of the graph, it indicates a positive correlation between the two variables. On the other hand, if the dots cluster around a straight line going from the top left to the bottom right of the graph, it indicates a negative correlation.

If the dots are spread out with no clear pattern, it suggests that the variables are not correlated. Scatter plots are also helpful in identifying any outliers or unusual data points that deviate significantly from the overall trend.

Overall, scatter plots are a powerful tool for understanding the relationship between two numerical variables and for making insights into the data's behavior. They are commonly used in data analysis, statistics, and various scientific fields to gain a visual understanding of data patterns and correlations.

```python

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,c='orange',alpha=0.9)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('scatter plot of x and y')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

ANSWER:
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple rows and columns of smaller axes (plots) and then plot different data or visualizations in each of these subplots. This feature is particularly useful when you want to compare or display multiple related plots side by side or in a structured layout.

The subplot() function takes three integer arguments: nrows, ncols, and index, which define the grid's dimensions and the position of the current subplot within that grid.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('x1 vs y1')
plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)
plt.title('x2 vs y2')
plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)
plt.title('x3 vs y3')
plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
plt.title('x4 vs y4')
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart, is a type of data visualization that uses rectangular bars to represent data. The length or height of each bar is proportional to the data values it represents. Bar plots are commonly used to display and compare the values of different categories or groups of data.

The main purpose of a bar plot is to show the distribution and relative magnitudes of discrete data points or categorical variables. It is particularly useful when you want to:

1. Compare data across different categories: Bar plots allow you to visually compare the values of different categories side by side. This makes it easy to identify which category has the highest or lowest value and observe any significant differences between them.

2. Display frequency or count data: If you have data that falls into discrete categories and you want to visualize the count or frequency of each category, a bar plot is an excellent choice.

3. Present survey results or poll data: Bar plots are commonly used to present survey responses, poll results, or any other data that involves multiple response options or categories.

4. Visualize trends over time: In some cases, bar plots can also be used to show how data in different categories changes over time, though line plots are more commonly used for this purpose.

A typical bar plot consists of bars drawn parallel to the x-axis, with each bar representing a specific category or group. The height of each bar is directly proportional to the value it represents. The bars are separated from each other, allowing clear visual distinction between the categories.

Bar plots are straightforward to create using various data visualization libraries in Python, such as Matplotlib or Seaborn. They are widely used in various fields, including business, social sciences, market research, data analysis, and more, to gain insights into the distribution and comparison of categorical data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color='red')
plt.xlabel('companies')
plt.ylabel('profit')
plt.title('company vs profit')
```

```python
#horizantal bar plot
```

```python
plt.barh(company,profit)
plt.title('horizantal bar plot of companies vs profit')
plt.xlabel('profit')
plt.ylabel('company')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200

ANSWER:
    A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through five summary statistics. It provides a concise way to visualize the spread, skewness, and potential outliers in the data. Box plots are particularly useful for comparing data distributions across different groups or categories.

A box plot consists of several components:

1. Box: The box represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third quartile (Q3). It contains 50% of the data points and gives a sense of the data's central tendency and spread.

2. Whiskers: The whiskers extend from the box and indicate the range of the data within a certain factor (often 1.5 times the IQR). They represent the minimum and maximum values within this range, excluding any outliers.

3. Outliers: Individual data points that fall outside the whisker range are considered outliers and are represented as individual dots or points beyond the whiskers. Outliers can provide valuable insights into unusual or extreme data values.

The main use of box plots is to compare the distribution of numerical data across different categories or groups. By creating a box plot for each category, you can easily visualize the central tendency, spread, and skewness of the data within each group. This allows for quick identification of variations or differences between groups and the presence of outliers.

Box plots are especially effective when dealing with large datasets or multiple groups, as they provide a compact and informative summary of the data without showing each individual data point. They are commonly used in exploratory data analysis, statistical analysis, and data visualization in various fields such as finance, medicine, social sciences, and more.

In Python, you can create box plots using libraries like Matplotlib, Seaborn, or Pandas. These libraries provide simple functions to generate box plots from data frames or arrays, making it easy to visualize and compare data distributions across different categories.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```
