### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

A t-test and a z-test are both statistical hypothesis tests used to determine if there is a significant difference between the means of two groups. However, they are used in different situations based on the characteristics of the data and the assumptions that can be made.

**T-test:**
A t-test is used when the sample size is small (typically less than 30) and the population standard deviation is unknown. It is more appropriate when dealing with small samples because it takes into account the uncertainty introduced by using the sample standard deviation instead of the population standard deviation. There are two main types of t-tests: the independent samples t-test and the paired samples t-test.

**Example scenario for a t-test:**
Let's say you want to test whether there is a significant difference in the average test scores of two groups of students (Group A and Group B) who have undergone different teaching methods. Since you only have a small number of students in each group and you don't know the population standard deviation, you would use an independent samples t-test to compare the means of the two groups.

**Z-test:**
A z-test is used when the sample size is large (typically greater than 30) and the population standard deviation is known or can be reasonably estimated. It is an approximation of the t-test when the sample size is sufficiently large, and it relies on the standard normal distribution (z-distribution). The z-test is more powerful and accurate when the sample size is large.

**Example scenario for a z-test:**
Imagine you are studying the heights of a large population of adult males. You have access to a sample of 200 individuals with their heights and you know the population standard deviation of heights based on previous research. In this case, you could use a z-test to test whether the average height of your sample is significantly different from a known population mean height.

In summary, choose between a t-test and a z-test based on the sample size, whether the population standard deviation is known, and the assumptions that can be made about the data. Use a t-test for smaller samples with unknown population standard deviation, and use a z-test for larger samples with known or estimated population standard deviation.

### Q2: Differentiate between one-tailed and two-tailed tests

One-Tailed Test:
A one-tailed test, also known as a one-sided test, is a type of statistical hypothesis test in which the alternative hypothesis is formulated to detect a difference or relationship in one specific direction. In other words, it tests whether a parameter is significantly greater than or significantly less than a certain value, but not both. The critical region for the test is located on only one side of the distribution.

Example: Suppose you are testing whether a new drug increases the average response time in a task. Your null hypothesis (H0) could be that the drug has no effect on response time, and your alternative hypothesis (H1) would state that the drug increases response time. This is a one-tailed test because you are only interested in detecting an increase in response time.

Two-Tailed Test:
A two-tailed test, also known as a two-sided test, is a type of statistical hypothesis test in which the alternative hypothesis is formulated to detect a difference or relationship in either direction. It tests whether a parameter is significantly different from a certain value, without specifying whether it's greater or less. The critical region for the test is split between both sides of the distribution.

Example: Let's say you are investigating whether a coin is fair (50% chance of heads and 50% chance of tails). Your null hypothesis (H0) could be that the coin is fair, and your alternative hypothesis (H1) would state that the coin is not fair. In this case, you are interested in detecting any significant deviation from the expected 50% probability in either direction, making it a two-tailed test.

In summary, the main difference between one-tailed and two-tailed tests lies in the directionality of the alternative hypothesis and the corresponding critical regions. One-tailed tests focus on a specific direction of effect (either greater or smaller), while two-tailed tests are more general and detect differences in either direction. The choice between the two types of tests depends on the research question and the specific hypothesis being tested.

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Type 1 and Type 2 errors are concepts in hypothesis testing that relate to the incorrect decisions made during the testing process.

**Type 1 Error (False Positive):**
A Type 1 error occurs when the null hypothesis (H0) is rejected when it is actually true. In other words, you conclude that there is a significant effect or difference when, in reality, there is none. This error is also known as a false positive or alpha error and is denoted by the symbol "α."

Example Scenario (Type 1 Error):
Imagine you are conducting a clinical trial to test the effectiveness of a new drug. The null hypothesis (H0) states that the drug has no effect on the condition being treated. However, due to random variability or other factors, your analysis leads you to reject the null hypothesis and conclude that the drug is effective, when in fact it is not. This would be a Type 1 error.

**Type 2 Error (False Negative):**
A Type 2 error occurs when the null hypothesis (H0) is not rejected when it is actually false. In other words, you fail to detect a significant effect or difference that truly exists. This error is also known as a false negative or beta error and is denoted by the symbol "β."

Example Scenario (Type 2 Error):
Suppose you are testing whether a new diagnostic test can accurately identify a certain disease. The null hypothesis (H0) states that the test is not effective in detecting the disease. However, the test actually does have the ability to identify the disease. Due to factors like small sample size or test limitations, your analysis fails to reject the null hypothesis, and you conclude that the test is not effective, when it actually is. This would be a Type 2 error.

It's important to note that there is a trade-off between Type 1 and Type 2 errors. As you decrease the risk of one type of error, the risk of the other type of error may increase. This trade-off is typically controlled by setting a significance level (alpha, α) for the test, which determines the threshold for rejecting the null hypothesis.

In summary, Type 1 errors involve falsely rejecting a true null hypothesis, while Type 2 errors involve failing to reject a false null hypothesis. Both types of errors are inherent in hypothesis testing and should be considered when interpreting the results of statistical analyses. The goal is to strike a balance between these errors based on the specific context and consequences of the test.

### Q4:  Explain Bayes's theorem with an example.

Bayes's theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis based on new evidence. It involves conditional probabilities and is often used in situations where we want to revise our beliefs about a hypothesis given observed data.

The formula for Bayes's theorem is as follows:

\[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

Where:
- \( P(A|B) \) is the probability of hypothesis \( A \) being true given that evidence \( B \) has been observed (the posterior probability).
- \( P(B|A) \) is the probability of observing evidence \( B \) given that hypothesis \( A \) is true (the likelihood).
- \( P(A) \) is the prior probability of hypothesis \( A \) being true before observing evidence \( B \).
- \( P(B) \) is the probability of observing evidence \( B \).

**Example: Medical Test**

Let's illustrate Bayes's theorem with a medical example. Consider a diagnostic test for a rare disease. The disease affects only 1% of the population. The test is not perfect, and it has a false positive rate of 5% (it indicates the disease is present when it's not) and a false negative rate of 2% (it indicates the disease is absent when it's actually present).

You want to calculate the probability that a person has the disease given that they tested positive for it.

Let:
- \( A \) be the event that the person has the disease.
- \( B \) be the event that the person tests positive for the disease.

We are interested in \( P(A|B) \), the probability that the person has the disease given that they tested positive.

Using Bayes's theorem:
\[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

- \( P(A) = 0.01 \) (prior probability of having the disease)
- \( P(B|A) = 1 - 0.02 = 0.98 \) (probability of testing positive if you have the disease)
- \( P(B|\neg A) = 0.05 \) (probability of testing positive if you don't have the disease)

Now, let's calculate \( P(B) \) using the law of total probability:
\[ P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A) \]
\[ P(B) = 0.98 \cdot 0.01 + 0.05 \cdot 0.99 \]

Now, we can calculate \( P(A|B) \):
\[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]
\[ P(A|B) = \frac{0.98 \cdot 0.01}{0.98 \cdot 0.01 + 0.05 \cdot 0.99} \]

After calculating this, you'll find that \( P(A|B) \) is approximately 0.165, which means that even if someone tests positive for the disease, the probability of them actually having it is still relatively low due to the rareness of the disease and the imperfect test.

This example demonstrates how Bayes's theorem can help update our beliefs based on new evidence and prior probabilities.

### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values within which we believe the true value of a population parameter lies, based on a sample from that population. It provides a measure of the uncertainty associated with estimating population parameters, such as means, proportions, or other statistical measures. The confidence interval is expressed as an interval estimate and is associated with a specified level of confidence, often expressed as a percentage.

**Calculating a Confidence Interval:**

To calculate a confidence interval, you typically follow these steps:

1. Collect a random sample from the population of interest.
2. Calculate the sample statistic of interest (e.g., sample mean or sample proportion).
3. Determine the level of confidence you want for the interval (e.g., 95%, 90%).
4. Use the appropriate formula or statistical tool to calculate the margin of error.
5. Construct the confidence interval by adding and subtracting the margin of error from the sample statistic.

The formula for a confidence interval is generally given by:

\[ \text{Confidence Interval} = \text{Sample Statistic} \pm \text{Margin of Error} \]

**Example:**

Suppose you are interested in estimating the average height of adult males in a certain city. You take a random sample of 100 adult males from the city and measure their heights. The sample mean height is 175 cm, and you want to calculate a 95% confidence interval for the true average height of all adult males in the city.

1. Sample Mean: \( \bar{x} = 175 \) cm
2. Sample Size: \( n = 100 \)
3. Confidence Level: 95% (which corresponds to a standard z-score of approximately 1.96 for a large sample)

The margin of error (\( \text{ME} \)) can be calculated using the formula:

\[ \text{ME} = \text{Critical Value} \times \frac{\text{Sample Standard Deviation}}{\sqrt{n}} \]

In this case, since the population standard deviation is unknown, we can use the sample standard deviation (\( s \)) as an estimate.

Suppose \( s = 10 \) cm.

\[ \text{ME} = 1.96 \times \frac{10}{\sqrt{100}} = 1.96 \]

Now, construct the confidence interval:

\[ \text{Confidence Interval} = \text{Sample Mean} \pm \text{Margin of Error} \]
\[ \text{Confidence Interval} = 175 \pm 1.96 \]

This yields a confidence interval of approximately \( (173.04, 176.96) \) cm.

Interpretation: We are 95% confident that the true average height of adult males in the city falls within the interval of 173.04 cm to 176.96 cm.

The confidence interval provides a range of heights within which we believe the population parameter (average height) is likely to lie, given the observed sample data and the specified confidence level. The wider the confidence interval, the greater the uncertainty associated with the estimation.

### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.

The z-score (also known as the standard score) measures how many standard deviations a data point is away from the mean of the population. It is calculated using the formula:

\[ z = \frac{x - \mu}{\sigma} \]

Where:
- \( z \) is the z-score.
- \( x \) is the data point value.
- \( \mu \) is the population mean.
- \( \sigma \) is the population standard deviation.

Given the values:
- Data Point Value (\( x \)) = 75
- Population Mean (\( \mu \)) = 70
- Population Standard Deviation (\( \sigma \)) = 5

Substitute these values into the formula to calculate the z-score:

\[ z = \frac{75 - 70}{5} = 1 \]

Interpretation: The z-score of 1 for the data point 75 indicates that this data point is 1 standard deviation above the mean of the population. In other words, the value 75 is higher than the average (mean) value of the population by 1 standard deviation. The positive sign of the z-score indicates that the data point is above the mean.

Z-scores are useful for comparing individual data points to the overall distribution and for identifying how extreme or typical a particular data point is within that distribution. Positive z-scores indicate values above the mean, negative z-scores indicate values below the mean, and a z-score of 0 indicates that the data point is exactly at the mean.

Now, calculate the confidence interval:
\[ \text{Confidence Interval} = \hat{p} \pm \text{Margin of Error} \]
